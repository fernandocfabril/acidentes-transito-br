### Importar bibliotecas

```python
# importar bibliotecas
import pandas as pd
```

### Importar arquivos CSV

```python
# importar o arquivo CSV
nome_arquivo = 'dados/2021.csv'

df2021 = pd.read_csv(nome_arquivo, sep=';', encoding='cp1252')
df2021
```

```python
# importar o arquivo CSV
nome_arquivo = 'dados/2022.csv'

df2022 = pd.read_csv(nome_arquivo, sep=',', encoding='utf-8')
df2022
```

```python
# importar o arquivo CSV
nome_arquivo = 'dados/2023.csv'

df2023 = pd.read_csv(nome_arquivo, sep=';', encoding='cp1252')
df2023
```

```python
# importar o arquivo CSV
nome_arquivo = 'dados/2024.csv'

df2024 = pd.read_csv(nome_arquivo, sep=',', encoding='utf-8')
df2024
```

```python
# concatenar os DataFrames
df = pd.concat([df2021, df2022, df2023, df2024], ignore_index=True)
df.shape
```

```python
# salvar o DataFrame concatenado em um novo arquivo CSV
df.to_csv('dados/acidentes_transito.csv', sep=';', encoding='utf-8', index=False)
```
