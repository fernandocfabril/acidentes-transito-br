As coluna que estão com valores nulos são:
`uop`
`delegacia`
`Regional`
`classificacao_acidente`

São todas colunas Categóricas (texto).

Vamos utilizar a MODA para substituir os valores nulos em cada coluna.

```python
# verificar se existe algum valor ausente
df.isnull().sum().sort_values(ascending=False)
```

```python
# mostrar a moda de cada coluna
df[['uop', 'delegacia', 'regional', 'classificacao_acidente']].mode()
```

```python
# substituir os valores nulos pela moda da coluna uop
df['uop'] = df['uop'].fillna(df['uop'].mode()[0])
```

```python
# substituir os valores nulos pela moda da coluna delegacia
df['delegacia'] = df['delegacia'].fillna(df['delegacia'].mode()[0])
```

```python
# substituir os valores nuloes plea moda da coluna regional
df['regional'] = df['regional'].fillna(df['regional'].mode()[0])
```

```python
# substituir os valores nulos pela moda da coluna classificacao_acidente
df['classificacao_acidente'] = df['classificacao_acidente'].fillna(df['classificacao_acidente'].mode()[0])
```

```python
# verificar se existe algum valor ausente (nulo)
df.isnull().sum().sort_values(ascending=False)
```

Algumas colunas que são do tipo `object` estão com tipo errado:

`data_inversa` deve ser do tipo DateTime\
`km` deve ser do tipo Float\
`latitude` deve ser do tipo Float\
`longitude` deve ser do tipo Float

```python
# mostrar todoas as colunas que são do tipo objet
df[df.select_dtypes(include='object').columns.tolist()]
```

```python
# alterar o tipo da coluna data_inversa de object para datatime
df['data_inversa'] = pd.to_datetime(df['data_inversa'], format='%Y-%m-%d')
```

```python
# alterar o tipo da coluna km de ojbect para float
df['km'] = df['km'].str.replace(',','.').astype(float)
```

```python
# alterar o tipo de coluna latitude de object para float
df['latitude'] = df['latitude'].str.replace(',','.').astype(float)
```

```python
# alterar o tipo de coluna longitude de object para float
df['longitude'] = df['longitude'].str.replace(',','.').astype(float)
```

```python
# mostrar todoas as colunas que são do tipo objet após as alterações
df[df.select_dtypes(include='object').columns.tolist()]
```

```python
df
```

```python
# criar uma coluna com o ano e o mês do acidente
df['ANO'] = df['data_inversa'].dt.year
df['MES'] = df['data_inversa'].dt.month
df[['data_inversa', 'ANO', 'MES']].head()
```

```python
# criar uma função para determinar o período do dia base na hora do acidente
def periodo_do_dia(hora):
    if hora < 6:
        return 'Madrugada'
    elif hora < 12:
        return 'Manhã'
    elif hora < 18:
        return 'Tarde'
    else:
        return 'Noite'
```

```python
# criar uma coluna com a hora do acidente
df['HORA'] = df['horario'].str.split(':').str[0].astype(int)
```

```python
# criar uma nova coluna com o período do dia, aplicando a função periodo_do_dia
df['PERIODO_DIA'] = df['HORA'].apply(periodo_do_dia)
```

```python
# mostrar as colunas criadas com o horario
df[['horario', 'HORA', 'PERIODO_DIA']]
```

[Veículos engavetam em grave acidente na BR-277, em Balsa Nova; PRF diz que pelo menos cinco pessoas morreram](https://g1.globo.com/pr/parana/noticia/2023/09/02/grave-acidente-br-277.ghtml)\

```python
# plotar um gráfico de boxplot com as principais variáveis
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['pessoas', 'mortos', 'feridos_leves', 'feridos_graves', 'ilesos', 'ignorados', 'feridos', 'veiculos']], orient='v')
plt.title('Bloxpot das principais variáveis')
plt.xlabel('Variáveis')
plt.ylabel('Valores')
plt.show()
```
